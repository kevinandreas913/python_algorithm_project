```python
# Import library tambahan jika diperlukan
import numpy as np  
import pandas as pd 
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_absolute_error, mean_squared_error
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Dropout
from sklearn.preprocessing import MinMaxScaler

# Bagian 1: Model ARIMA untuk prediksi awal dan ekstraksi residual
# Load dataset
df = pd.read_csv("Superstore.csv", sep=",") 
sales_data = df[['Order Date', 'Sales']] 
sales_data['Order Date'] = pd.to_datetime(sales_data['Order Date'])
sales_data = sales_data.set_index('Order Date')
monthly_sales = sales_data.resample('M').mean()

# Check stationarity
def check_stationarity(timeseries): 
    from statsmodels.tsa.stattools import adfuller 
    result = adfuller(timeseries, autolag='AIC') 
    p_value = result[1]
    if p_value < 0.05:
        print("Data stasioner")
    else:
        print("Data tidak stasioner")

check_stationarity(monthly_sales['Sales'])

# Fit ARIMA model
p, d, q = 12, 2, 1
model_arima = ARIMA(monthly_sales, order=(p, d, q))
results_arima = model_arima.fit()

# Forecast ARIMA
Periode = 12
forecast_arima = results_arima.get_forecast(steps=Periode)
pred_arima = forecast_arima.predicted_mean

# Calculate residuals between actual and ARIMA predicted values
residuals = monthly_sales['Sales'] - results_arima.predict(start=0, end=len(monthly_sales)-1)

# Bagian 2: Menggunakan LSTM untuk memprediksi residuals
# Scaling residual data
scaler = MinMaxScaler(feature_range=(0, 1))
residuals_scaled = scaler.fit_transform(residuals.values.reshape(-1, 1))

# Convert residuals to a supervised problem for LSTM
def create_dataset(data, look_back=1):
    X, Y = [], []
    for i in range(len(data)-look_back):
        X.append(data[i:i+look_back, 0])
        Y.append(data[i + look_back, 0])
    return np.array(X), np.array(Y)

look_back = 12
trainX, trainY = create_dataset(residuals_scaled, look_back)
trainX = np.reshape(trainX, (trainX.shape[0], 1, trainX.shape[1]))

# Build LSTM model
model_lstm = Sequential()
model_lstm.add(LSTM(50, return_sequences=True, input_shape=(1, look_back)))
model_lstm.add(Dropout(0.2))
model_lstm.add(LSTM(50))
model_lstm.add(Dropout(0.2))
model_lstm.add(Dense(1))
model_lstm.compile(loss='mean_squared_error', optimizer='adam')

# Train LSTM on residuals
model_lstm.fit(trainX, trainY, epochs=50, batch_size=16, verbose=1)

# Predict residuals using LSTM
residual_forecast_lstm = []
current_input = trainX[-1]  # Last training sample as starting point for forecast

for _ in range(Periode):
    lstm_pred = model_lstm.predict(np.reshape(current_input, (1, 1, look_back)))
    residual_forecast_lstm.append(lstm_pred[0, 0])
    current_input = np.roll(current_input, -1)
    current_input[0, -1] = lstm_pred

# Inverse transform predictions to original scale
residual_forecast_lstm = scaler.inverse_transform(np.array(residual_forecast_lstm).reshape(-1, 1)).flatten()

# Bagian 3: Menggabungkan prediksi ARIMA dan LSTM
hybrid_forecast = pred_arima.values + residual_forecast_lstm

# Plot hasil prediksi hybrid model
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.plot(monthly_sales.index, monthly_sales['Sales'], label='Actual Sales', color='black')
plt.plot(pred_arima.index, pred_arima, label='ARIMA Forecast', color='blue')
plt.plot(pred_arima.index, hybrid_forecast, label='Hybrid ARIMA-LSTM Forecast', color='red')
plt.xlabel("Order Date")
plt.ylabel("Sales")
plt.title("Sales Forecast: Hybrid ARIMA-LSTM Model")
plt.legend()
plt.show()

# Evaluate the hybrid model
mae = mean_absolute_error(monthly_sales[-Periode:], hybrid_forecast)
mse = mean_squared_error(monthly_sales[-Periode:], hybrid_forecast)
print(f'Hybrid Model MAE: {mae}')
print(f'Hybrid Model MSE: {mse}')
```
