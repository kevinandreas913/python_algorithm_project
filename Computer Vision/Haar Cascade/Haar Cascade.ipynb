# Algoritma Haar Cascade untuk Deteksi Jumlah Orang

Algorima Haar Cascade -> algoritma untuk mendeteksi wajah. 

### Algoritma Haar Cascade menggunakan fungsi wavalet haar.

- Haar Wavelet adalah teknik dekomposisi sinyal yang digunakan untuk menganalisis dan memproses sinyal.
- Haar Cascade adalah algoritma klasifikasi yang digunakan untuk mendeteksi objek dalam gambar.

# Contoh ide dari Haar Cascade



```python
import cv2 as cv
from matplotlib import pyplot as plt
```

```python
nama = input("Masukkan nama file gambar Anda yang telah disimpan dalam file jumlahorang (misalnya 'gambar.jpg'): ")
path_gambar = 'HaarCascade/' + nama
foto = cv.imread(path_gambar)
if foto is None:
    raise IOError('Gambar tidak dapat ditemukan atau dibuka')
else:
    print("Foto ditemukan!")
```

### Directori: 

```python
import cv2
print(cv2.__file__)
```

```python
cc_wajah = cv.CascadeClassifier('HaarCascade/haarcascade_frontalface_alt.xml')
if cc_wajah.empty():
    raise IOError('File XML tidak bisa ditemukan')
else:
    print("File ditemukan!")
```

```python
foto_rgb = cv.cvtColor(foto, cv.COLOR_BGR2RGB)
plt.imshow(foto_rgb)
plt.axis('off') 
plt.show()
```

```python
plt.imshow(foto)
plt.axis('off') 
plt.show()
```

```python
foto_gray = cv.cvtColor(foto, cv.COLOR_BGR2GRAY)

hasil = cc_wajah.detectMultiScale(foto_gray, scaleFactor = 1.1, minNeighbors=3)
# hasil = cc_wajah.detectMultiScale(foto_gray, scaleFactor = 1.2, minNeighbors=1)
print("Terdeteksi ", len(hasil), " objek di dalam foto.")
for (x,y,w,h) in hasil:
    cv.rectangle(foto, (x,y), (x+w,y+h), color = (0, 255, 0), thickness=3)
# cv.imshow("Deteksi Wajah", foto)
# cv.imwrite('jumlahorang/hasil_deteksi.jpg', foto)
# cv.waitKey(0)
# cv.destroyAllWindows()
foto_rgb = cv.cvtColor(foto, cv.COLOR_BGR2RGB)
plt.imshow(foto_rgb)
plt.axis('off') 
plt.show()
```
