- Hebb Neural Network
- Backpopragation

```python
import pandas as pd
import sklearn.neural_network as ann
import sklearn.model_selection as ms

try:
    df_hasil = pd.read_csv('MLP.csv', sep=";")
    print("Pembacaan data berhasil")
except FileNotFoundError:
    print("Pembacaan file gagal")
print("Data adalah:")
display(df_hasil)
```

```python
X = df_hasil.drop(['Kode', 'Hasil'], axis=1)
y = df_hasil['Hasil']
display(X)
X.describe()
```

```python
X_train, X_test, y_train, y_test = ms.train_test_split(X, y, test_size=0.2)
X_train.count()
```

```python
import sklearn.preprocessing as pp
scl = pp.StandardScaler(copy=True, with_mean=True, with_std=True)
scl.fit(X_train)

# Transformasi data latih dan data uji
X_train = scl.transform(X_train)
X_test = scl.transform(X_test)
```

```python
print(X_train.min())
print(X_train.max())
```

```python
mip = ann.MLPClassifier(hidden_layer_sizes=(3), max_iter = 5)
mip.fit(X_train, y_train)
```

```python
y_prediksi = mip.predict(X_test)
print(y_prediksi)
```

```python
import sklearn.metrics as met
print(met.classification_report(y_test, y_prediksi))
```

```python
data_baru = pd.DataFrame({
    'Prestasi': [90, 60, 85],
    'Tanggungan': [4, 3, 2],
    'Listrik': [220000, 300000, 150000]
})
data_baru_scaled = scl.transform(data_baru)
hasil_prediksi = mip.predict(data_baru_scaled)
for i, pred in enumerate(hasil_prediksi):
    if pred == 1:
        keterangan = "Menerima Beasiswa"
    else:
        keterangan = "Gagal Menerima Beasiswa"
    print(f"Prediksi untuk data baru {i+1}: {pred} ({keterangan})")
```
